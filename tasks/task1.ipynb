Problem1 :  
Plot a sin function as well as noise corrupted data from this sin function .

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2 * np.pi, 100)
y_clean = np.sin(x)
noise = np.random.normal(0, 0.2, size=x.shape)
y_noisy = y_clean + noise
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x, y_clean, label='sin fun', linewidth=2)
plt.scatter(x, y_noisy, label= 'noise', color='red', alpha=0.6, s=20)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.grid(True)
plt.show()
```

Problem 2:  
Determine if an array is monotonic. (Without using Python function)

```python
def is_monotonic(arr):
    increasing = True
    decreasing = True

    for i in range(1, len(arr)):
        if arr[i] > arr[i - 1]:
            decreasing = False
        elif arr[i] < arr[i - 1]:
            increasing = False

    return increasing or decreasing
```

```python

print(is_monotonic([1,2,3,3,3,3,3,4,5,6,7]))   
print(is_monotonic([7,6,5,4,3,3,3,3,3,2,1]))        
print(is_monotonic([1,5,4,3,3,3,3,3,2,6,7]))         
```
